# Islam figure 3 (similar to Jirka figure 3)

```python
def myFunction(f):
    F2s = f[0]
    h1 = f[1]
    h2 = f[2]
    q1 = f[3]
    q2 = f[4]
    
    F = np.zeros(5)
    F[0] = F1s - (1-Drho/rho_2*F1s)*(1-F2s)
    F[1] = Drho/rho_2*F1s - (H1/h1 + H2/h1 - h2/h1 - 1)
    F[2] = F2s/2 - F1s/2*h1/h2 + 1 - H2/h2
    F[3] = F1s - pow(q1,2)/(gp*pow(h1,3))
    F[4] = F2s - pow(q2,2)/(gp*pow(h2,3))
    
    return F
```

```python
g = 9.81 # [m s^-2] gravitatieversnelling
rho_1 = 1000 # [kg m^-3] dichtheid zoet water
rho_2 = 1022 # [kg m^-3] dichtheid zout water
Drho = rho_2 - rho_1 # [kg m^-3] dichtheidsverschil
gp = g*Drho / (rho_2+rho_1)
H_B = -17 # [mNAP] niveau bodem
H_Z = 2 # [mNAP] waterniveau zeezijde
H_M = 3 # [mNAP] waterniveau meerzijde

H_T = H_M - H_B

H2_arr = np.array([5, 6.5, 10, 12.5, 15, 18])
plotcolor = np.array(['m', 'r', 'g', 'b', 'y', 'k'])

plt.figure(figsize=(8,6))
plt.axis([0,2,0,1])
plt.xlabel('$F_l^2$')
plt.ylabel('$\lambda$')

#warnings.filterwarnings('ignore')
for i in range(H2_arr.size):
    H2 = H2_arr[i]
    H1 = H_T - H2
    F1ss = np.linspace(0,1,1000)
    Fls = np.zeros((F1ss.size))
    lam = np.zeros((F1ss.size))
    for ii in range(F1ss.size):
        F1s = F1ss[ii]
        fGuess = np.array([1,1,1,1,1])
        F2s, h1, h2, q1, q2 = fsolve(myFunction, fGuess)
        Fls[ii] = pow((q1+q2),2)/(gp*pow(H2,3))
        lam[ii] = q1/(q1+q2)

    plt.plot(Fls**.5, lam, plotcolor[i], label='{}'.format(H_T/H2))

lgnd = plt.legend(title='$H_T/H_2$:')
[lgnd.legendHandles[jj]._legmarker.set_markersize(10) for jj in range(6)]


H2_arr = np.linspace(1,20,100)
Flmax = np.zeros(H2_arr.size)
lammax = np.zeros(H2_arr.size)
for kk in range(H2_arr.size):
    H2 = H2_arr[kk]
    H1 = H_T - H2
    F1ss = np.linspace(0,1,1000)
    Fls = np.zeros((F1ss.size))
    lam = np.zeros((F1ss.size))
    for ii in range(F1ss.size):
        F1s = F1ss[ii]
        fGuess = np.array([1,1,1,1,1])
        F2s, h1, h2, q1, q2 = fsolve(myFunction, fGuess)
        Fls[ii] = pow((q1+q2),2)/(gp*pow(H2,3))
        lam[ii] = q1/(q1+q2)

    #print(Fls[Fls**.5==max(Fls**.5)]**.5)
    Flmax[kk] = Fls[Fls**.5==max(Fls**.5)]**.5
    lammax[kk] = lam[Fls**.5==max(Fls**.5)]

plt.plot(Flmax, lammax, '--r')
#plt.savefig('figures/discharge_ratios.png')
```

# Jirka fig 3

```python
g = 9.81
rho_1 = 1000 # [kg m^-3] dichtheid zoet water
rho_2 = 1022 # [kg m^-3] dichtheid zout water
Drho = rho_2 - rho_1 # [kg m^-3] dichtheidsverschil
gp = g*Drho / (rho_2+rho_1)

H = 20
hr = 0.7*H
```

```python
def myFunction(f):
    q1 = f[0]
    q2 = f[1]
    hc = f[2]
    
    F = np.zeros(3)

    F[0] = pow(q1,2)/(gp*pow(H-hc,3)) + pow(q2,2)/(gp*pow(hc,3)) - 1
    F[1] = hr - hc - pow(q2,2)/(2*gp*pow(hc,2)) + pow(q1,2)/(2*gp*pow(H-hc,2))
    F[2] = q - q1 - q2
    
    return F
```

```python
cols = np.array(['.r', '.g', '.y', '.b', '.m'])
colsi = 0
for hr in np.array([0.1*H, 0.3*H, 0.5*H, 0.8*H, H]):
    for q in np.linspace(0.1,50,10000):
        fGuess = np.array([1,1,2])
        q1, q2, hc = fsolve(myFunction, fGuess)
        plt.plot(q/np.sqrt(gp*hr**3), q1/q, cols[colsi], markersize=1)
    colsi += 1
plt.xlim([0,2])
plt.ylim([0,0.7])
```
